```python
path = '/content/drive/MyDrive/Colab Notebooks/Computer Vision/Face Detection Using Image/XML Files/haarcascade_frontalface_default.xml'
eye_path = '/content/drive/MyDrive/Colab Notebooks/Computer Vision/Face Detection Using Image/XML Files/haarcascade_eye.xml'
img_path = '/content/drive/MyDrive/Colab Notebooks/Computer Vision/Face Detection Using Image/Input Images/friends.jpeg'
img_path2 = '/content/drive/MyDrive/Colab Notebooks/Computer Vision/Face Detection Using Image/Input Images/GOPR1596.JPG'
```

```python
face=cv2.CascadeClassifier(path) #for detecting face
eye = cv2.CascadeClassifier(eye_path) #for detecting eyes
```

```python
image=cv2.imread(img_path2)
gray= cv2.cvtColor(image,cv2.COLOR_BGR2GRAY) #convert into gray 
```

```python
cv2_imshow(gray)
```

```python
#parameters(img,scale_factor[reduce image size],min_neighbour)
faces = face.detectMultiScale(gray,4,4)   #for  faces
```

```python
for(x,y,w,h) in faces:
    
    image=cv2.rectangle(image,(x,y),(x+w,y+h),(127,0,205),3)
    
    #Now detect eyes
    roi_gray = gray[y:y+h, x:x+w]
    roi_color = image[y:y+h, x:x+w]
    eyes = eye.detectMultiScale(roi_gray,1.2,1)
    for (ex,ey,ew,eh) in eyes:
        cv2.rectangle(roi_color,(ex,ey),(ex+ew,ey+eh),(255,0,0),2)
```
